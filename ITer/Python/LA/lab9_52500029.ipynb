```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def Find_lst_sq_sol(A,b):
    ATA = np.transpose(A)@np.array(A)
    ATb = np.transpose(A)@np.array(b)
    ATA_inv = np.round(np.linalg.inv(ATA),5)
    res = ATA_inv@ATb
    return res
```

```python
#main 
#1
A = np.array([[2,2],[2,3]])
b = np.array([[4],[6]])
print(Find_lst_sq_sol(A,b))
```

```python
# 2
A = np.array([
    [0, 0, 1],
    [0, 1, 1],
    [1, 2, 1],
    [1, 0, 1],
    [4, 1, 1],
    [4, 2, 1]
])
b = np.array([
    [0.5],
    [1.6],
    [2.8],
    [0.8],
    [5.1],
    [5.9]
])
print(Find_lst_sq_sol(A,b))
```

```python
# cau 4
x4 = np.array([2000, 6000, 20000, 30000, 40000])
y4 = np.array([20, 18, 10, 6, 2])
A4 = np.vstack([x4, np.ones(len(x4))]).T
w4 = np.linalg.lstsq(A4, y4, rcond=None)[0]
print(f"Cau 4: y = {w4[0]:.5f}x + {w4[1]:.5f}")

plt.scatter(x4, y4, c='r')
plt.plot(x4, w4[0]*x4 + w4[1], c='b')
plt.title("Cau 4")
plt.show()
```

```python
# cau 5
x5 = np.array([1, 2, 3])
y5 = np.array([7.9, 5.4, -9])
A5 = np.vstack([np.cos(x5), np.sin(x5)]).T
w5 = np.linalg.lstsq(A5, y5, rcond=None)[0]
print(f"Cau 5: y = {w5[0]:.5f}*cos(x) + {w5[1]:.5f}*sin(x)")
```

```python
# cau 6
x6 = np.array([1, 2, 3, 4, 5, 6])
y6 = np.array([2.1, 3.5, 4.2, 3.1, 4.4, 6.8])
A6 = np.vstack([x6**3, x6**2, x6, np.ones(len(x6))]).T
w6 = np.linalg.lstsq(A6, y6, rcond=None)[0]
print(f"Cau 6: y = {w6[0]:.5f}x^3 + {w6[1]:.5f}x^2 + {w6[2]:.5f}x + {w6[3]:.5f}")
plt.scatter(x6, y6, c='r')
t6 = np.linspace(1, 6, 100)
plt.plot(t6, w6[0]*t6**3 + w6[1]*t6**2 + w6[2]*t6 + w6[3], c='b')
plt.title("Cau 6")
plt.show()
```

```python
def scaling_draw(lam,miu,pts):
    S=np.array([[lam,0],[0,miu]])
    trans=S@pts.T
    plt.plot(pts[:,0],pts[:,1])
    plt.plot(trans[0],trans[1])
```

```python
pts = np.array([
    [0,1],
    [2, 1],
    [4,5],
    [0,1]
    
])
scaling_draw(2,2,pts)
```

```python
scaling_draw(0.5,0.5,pts)
```

```python
scaling_draw(1,-1,pts)
```

```python
scaling_draw(-1,1,pts)
```

```python
# cau 8
def ex8(theta, name):
    R = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]])
    pts = np.array([[0, 1, 1, 0, 0], [0, 0, 1, 1, 0]])
    new_pts = R @ pts
    
    plt.plot(pts[0], pts[1], 'b--', )
    plt.plot(new_pts[0], new_pts[1], 'r-', )
    plt.title(f"Cau 8{name}")
    plt.axis('equal')
    plt.show()

print("Cau 8a: Quay goc pi (180 do)")
ex8(np.pi, 'a')

print("Cau 8b: Quay goc pi/3 (60 do)")
ex8(np.pi/3, 'b')
```
