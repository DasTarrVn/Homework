```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
def Find_lst_sq_sol(A,b):
    ATA = np.transpose(A)@np.array(A)
    ATb = np.transpose(A)@np.array(b)
    ATA_inv = np.round(np.linalg.inv(ATA),5)
    res = ATA_inv@ATb
    return res
```

```python
#main 
#1
A = np.array([[2,2],[2,3]])
b = np.array([[4],[6]])
print(Find_lst_sq_sol(A,b))
```

```python
# 2
A = np.array([
    [0, 0, 1],
    [0, 1, 1],
    [1, 2, 1],
    [1, 0, 1],
    [4, 1, 1],
    [4, 2, 1]
])
b = np.array([
    [0.5],
    [1.6],
    [2.8],
    [0.8],
    [5.1],
    [5.9]
])
print(Find_lst_sq_sol(A,b))
```

```python
# 3
# a
pts = np.array([[0,1],
               [1,1],
               [2,2],
               [3,2]])
b=pts[:,1:2]
A=np.array([[1,x] for x,y in pts])
print(Find_lst_sq_sol(A,b))
```

```python
def scaling_draw(lam,miu,pts):
    S=np.array([[lam,0],[0,miu]])
    trans=S@pts.T
    plt.plot(pts[:,0],pts[:,1])
    plt.plot(trans[0],trans[1])
```

```python
pts = np.array([
    [0,1],
    [2, 1],
    [4,5],
    [0,1]
    
])
scaling_draw(2,2,pts)
```

```python
scaling_draw(0.5,0.5,pts)
```

```python
scaling_draw(1,-1,pts)
```

```python
scaling_draw(-1,1,pts)
```
