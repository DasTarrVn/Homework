```python
import numpy as np
```

```python
def is_square(A):
    return A.shape[0]==A.shape[1]
```

```python
def is_symmetric(A):
    return np.array_equal(A,A.T)
```

```python
def is_skew_symmetric(A):
    return np.array_equal(A,(-A).T)
  
```

```python
def upper_tri(A):
    return np.triu(A)
```

```python
def lower_tri(A):
    return np.tril(A)
```

```python
A = np.array([[2,4,1],[6,7,2],[3,5,9]])
print("(a)\n",is_square(A))
print("(b)\n",is_symmetric(A))
print("(c)\n",is_skew_symmetric(A))
print("(d)\n",upper_tri(A))
print("(e)\n",lower_tri(A))
```

```python
A = np.array([
    [6, 0, 0, 5],
    [1, 7, 2, -5],
    [2, 0, 0, 0],
    [8, 3, 1, 8]
])
B = np.array([
    [1, -2, 5, 2],
    [0, 0, 3, 0],
    [2, -6, -7, 5],
    [5, 0, 4, 4]
])
C = np.array([
    [3, 5, -8, 4],
    [0, -2, 3, -7],
    [0, 0, 1, 5],
    [0, 0, 0, 2]
])
D = np.array([
    [4, 0, 0, 0],
    [7, -1, 0, 0],
    [2, 6, 3, 0],
    [5, -8, 3, 0],
    [5, -8, 4, -3]
])
E = np.array([
    [4, 0, -7, 3, -5],
    [0, 0, 2, 0, 0],
    [7, 3, -6, 4, -8],
    [5, 0, 5, 2, -3],
    [0, 0, 9, -1, 2]
])
F = np.array([
    [6, 3, 2, 4, 0],
    [9, 0, -4, 1, 0],
    [8, -5, 6, 7, 1],
    [3, 0, 0, 0, 0],
    [4, 2, 3, 2, 0]
])
arr={"A": A,"B":B,"C":C,"D":D,"E":E,"F":F}

for x in arr:
        try:
            tmp= np.linalg.det(arr[x])
            print(f"Det {x}: {tmp}")
        except Exception as e:
            print(e)


```

```python

A1, B1 = np.random.randint(-10, 10, (5, 5)), np.random.randint(-10, 10, (5, 5))
A2, B2 = np.random.randint(-10, 10, (2, 2)), np.random.randint(-10, 10, (2, 2))
A3, B3 = np.random.randint(-10, 10, (3, 3)), np.random.randint(-10, 10, (3, 3))
A4, B4 = np.random.randint(-10, 10, (4, 4)), np.random.randint(-10, 10, (4, 4))

arr13 = {"(5x5)": (A1, B1), "(2x2)": (A2, B2), "(3x3)": (A3, B3), "(4x4)": (A4, B4)}

for x in arr13:
    A, B = arr13[x]
    diff = np.linalg.det(A + B) - np.linalg.det(A) - np.linalg.det(B)
    print(f"{x} -> det(A+B) - detA - detB = {diff:.2f}")
print("=> SAI.")


arr14 = {
    "(2x2)": (np.random.randint(-5, 5, (2, 2)), np.random.randint(-5, 5, (2, 2))),
    "(3x3)": (np.random.randint(-5, 5, (3, 3)), np.random.randint(-5, 5, (3, 3))),
    "(4x4)": (np.random.randint(-5, 5, (4, 4)), np.random.randint(-5, 5, (4, 4))),
    "(5x5)": (np.random.randint(-5, 5, (5, 5)), np.random.randint(-5, 5, (5, 5)))
}

for x in arr14:
    A, B = arr14[x]
    det_AB = np.linalg.det(np.matmul(A, B))
    detA_detB = np.linalg.det(A) * np.linalg.det(B)
    diff = det_AB - detA_detB
    print(f"{x} -> det(AB) - (detA * detB) = {diff:.2f}")
print("=> ĐÚNG.")
```

```python

A = np.array([
    [2, 4, 5/2],
    [-3/4, 2, 1/4],
    [1/4, 1/2, 2]
])

B = np.array([
    [1, -1/2, 3/4],
    [3/2, 1/2, -2],
    [1/4, 1, 1/2]
])
A_1 = np.linalg.inv(A)
B_1 = np.linalg.inv(B)
print("(a)")
print("A^-1*B^-1\n",A_1@B_1)
A_nhan_B =  A@B
print("(A*B)^-1\n",np.linalg.inv(A_nhan_B))
B_nhan_A = B@A
print("(B*A)^-1\n",np.linalg.inv(B_nhan_A))
print("(b)")
print("(A^-1)^T\n",A_1.T)
print("(A^T)^-1\n",np.linalg.inv(A.T))
```
