```python
#Ex1:
def is_linear_combination(A,w):
    A_=np.column_stack((A,w))
    R, pivot_index = sp.Matrix(A_).rref()
    return pivot_index[-1]<A_.shape[1]-1
#Ex2:
def KT_DLTT(A):
    R, pivot_index= sp.Matrix(A).rref()
    return len(pivot_index) == len(A.T)
```

```python
A = sp.Matrix([
    [1, 0, 2, 3],
    [4, -1, 0, 2],
    [0, -1, -8, -10]
])
R, pivots = A.rref()
print("RREF:")
sp.pprint(R)
n = A.shape[1]
free_vars = [i for i in range(n) if i not in pivots]
basis = []
for j in free_vars:
    x = sp.zeros(n, 1)
    x[j] = 1
    for i, pivot_col in enumerate(pivots):
        x[pivot_col] = -R[i, j]
    basis.append(x)
    print("\nVector nghiệm:")
    sp.pprint(x)
print("\nKiểm tra A * v = 0:")
for v in basis:
    sp.pprint(A * v)
a, b = sp.symbols('a b')
v1, v2 = basis
LC = a*v1 - b*v2
print("\nA * (a*v1 - b*v2) =")
sp.pprint(A * LC)
```

```python
def check_vector(A, w):
    print("Matrix A:")
    sp.pprint(A)
    print("\nVector w:")
    sp.pprint(w)
    aug = A.row_join(w)  
    R, pivots = aug.rref()
    print("RREF của [A|w]:")
    sp.pprint(R)
    m, n = A.shape
    has_solution = True
    for i in range(m):
        if all(R[i, j] == 0 for j in range(n)) and R[i, n] != 0:
            has_solution = False
    if has_solution:
        print("=> w thuộc Col(A)")
    else:
        print("=> w không thuộc Col(A)")
    result = A * w
    sp.pprint(result)
    if result == sp.zeros(A.shape[0], 1):
        print("=> w thuộc Null(A)")
    else:
        print("=> w không thuộc Null(A)")
    print("\nKẾT LUẬN:")
    if has_solution and result == sp.zeros(A.shape[0], 1):
        print("=> w thuộc CẢ HAI")
    elif has_solution:
        print("=> w thuộc Column space")
    elif result == sp.zeros(A.shape[0], 1):
        print("=> w thuộc Null space")
    else:
        print("=> w KHÔNG thuộc cả hai")
    print("\n" + "="*50 + "\n")
# (a)
A1 = sp.Matrix([
    [7, 6, -4, 1],
    [-5, -1, 0, -2],
    [9, -11, 7, -3],
    [19, -9, 7, 1]
])
w1 = sp.Matrix([1, 1, -1, -3])
check_vector(A1, w1)
# (b)
A2 = sp.Matrix([
    [-8, 5, -2, 0],
    [-5, 2, 1, -2],
    [10, -8, 6, -3],
    [3, -2, 1, 0]
])
w2 = sp.Matrix([1, 2, 1, 0])
check_vector(A2, w2)
```

```python
def is_orthogonal_set(vectors):
    n = len(vectors)
    for i in range(n):
        for j in range(i+1, n):
            dot = vectors[i].dot(vectors[j])
            print(f"<u{i+1}, u{j+1}> =", dot)
            if dot != 0:
                print("=> Không trực giao")
                return False
    print("=> Tập trực giao")
    return True
u1 = sp.Matrix([3, 1, 1])
u2 = sp.Matrix([-1, 2, 1])
u3 = sp.Matrix([-1/2, 2, 7/2])
vectors = [u1, u2, u3]
is_orthogonal_set(vectors)
```

```python
def projection(y, u):
    proj = (y.dot(u) / u.dot(u)) * u
    print("Projection của y lên u:")
    sp.pprint(proj)
    return proj
y = sp.Matrix([7, 6])
u = sp.Matrix([4, 2])
projection(y, u)
```

```python
def is_orthonormal(U):
    print("Matrix U:")
    sp.pprint(U)
    result = U.T * U
    print("\nU^T U =")
    sp.pprint(result)
    I = sp.eye(U.shape[1])
    if result == I:
        print("=> Có orthonormal columns")
    else:
        print("=> KHÔNG orthonormal")
U = sp.Matrix([
    [1/sp.sqrt(2), 1/sp.sqrt(2)],
    [1/sp.sqrt(2), -1/sp.sqrt(2)]
])
is_orthonormal(U)
```

```python
def gram_schmidt(A):
    cols = A.shape[1]
    U = []
    for i in range(cols):
        v = A[:, i]
        for u in U:
            proj = (v.dot(u) / u.dot(u)) * u
            v = v - proj
        U.append(v)
    print("Orthogonal basis:")
    for u in U:
        sp.pprint(u)
    return U
A = sp.Matrix([
    [-10, 13, 7, -11],
    [2, 1, -5, 3],
    [-6, 3, 13, -3],
    [16, -16, -2, 5],
    [2, 1, -5, -7]
])
gram_schmidt(A)
```
