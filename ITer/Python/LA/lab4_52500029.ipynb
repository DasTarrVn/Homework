```python
T = np.array([[0.6, 0.7],
              [0.4, 0.3]])

p = np.array([0.5, 0.5])

k_values = [1, 2, 10, 100, 100000]

for k in k_values:
    Tk = np.linalg.matrix_power(T, k)
    
    pk = Tk @ p
    
    print(k,": ",pk)
```

```python
A = np.array([[-1,4,8],
              [-9,1,2]]) 
B = np.array([[5,8],
              [0,-6],
              [5,6]])
C = np.array([[-4,1],
              [6,5]])
D = np.array([[-6,3,1],
              [8,9,-2],
              [6,-1,5]])
# a = A @ B.T

# (b)
b = B @ C.T
print("(b)\n", b)

# (c)
c = C - C.T
print("(c)\n", c)

# (d)
d = D - D.T
print("(d)\n", d)

# (e)
e = (D.T).T
print("(e)\n", e)

# (f)
f = 2 * C.T
print("(f)\n", f)

# (g)
g = A.T + B
print("(g)\n", g)

# (h)
h = (A.T + B).T
print("(h)\n", h)

# (i)
i = (2 * A.T - 5 * B).T
print("(i)\n", i)

# (j)
j = (-D).T
print("(j)\n", j)

# (k)
k_res = -(D.T)
print("(k)\n", k_res)

l = np.linalg.matrix_power(C, 2).T
print("(l)\n", l)

# (m)
m = np.linalg.matrix_power(C.T, 2)
print("(m)\n", m)
```
