```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
#(a)
x, y, z ,t = sp.symbols('x y z t')
eqa1 = sp.Eq(x+2*y+z,0)
eqa2 = sp.Eq(2*x-y+z,0)
eqa3 = sp.Eq(2*x+y,0)
solution_a= sp.solve([eqa1,eqa2,eqa3],[x,y,z])
print("(a)",solution_a)
#(b)

eqb1 = sp.Eq(2*x+y+z+t,1)
eqb2 = sp.Eq(x+2*y+z+t,1)
eqb3 = sp.Eq(x+y+2*z+2*t,1)
eqb4 = sp.Eq(x+y+z+2*t,1)
solution_b= sp.solve([eqb1,eqb2,eqb3,eqb4],[x,y,z,t])
print("(b)",solution_b)

```

```python
def draw2d(a,b,c,color):
    x_arr = np.arange(-3,3,0.1)
    if b!=0:
        y = lambda x: -a/b * x + c/b
        y_arr=list(map(y,x_arr))
        plt.plot(x_arr,y_arr,color = color, label = f'{a}x + {b}y = {c}')
    else:
        if a!= 0:
            y_arr = x_arr.copy()
            x_arr = np.full_like(y_arr,c/a)
            plt.plot(x_arr,y_arr,color = color, label = f'{a}x + {b}y = {c}')
        else:
            print("Khong ve duoc")
```

```python
draw2d(0, 1, -2, "red")   
draw2d(0, 1, -1, "blue")

plt.grid()   
plt.legend()  
plt.show()   
```

```python
draw2d(1, 1, 2, "red")

draw2d(1, -1, 0, "blue")
plt.grid()
plt.legend()
plt.show()
```

```python
draw2d(1, 1, 2, "red")
draw2d(2, 2, 4, "blue")
plt.grid()
plt.legend()
plt.show()
```

```python
def draw3d(a, b, c, d, color, ax):
    point_arr = np.arange(-5, 5, 0.5)
    
    if c != 0:
        X, Y = np.meshgrid(point_arr, point_arr)
        Z = (d - a*X - b*Y) / c
        ax.plot_surface(X, Y, Z, color=color, alpha=0.5)
        
    else:
        if b != 0:
            X, Z = np.meshgrid(point_arr, point_arr)
            Y = (d - a*X) / b
            ax.plot_surface(X, Y, Z, color=color, alpha=0.5)
            
        else:
            if a != 0:
                Y, Z = np.meshgrid(point_arr, point_arr)
                X = np.full_like(Y, d/a, dtype=float)
                ax.plot_surface(X, Y, Z, color=color, alpha=0.5)
                
            else:
                print("Khong ve duoc")
```

```python
#Tạo khung
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

draw3d(1, 1, 1, 1, 'red', ax)
draw3d(1, 1, 1, 2, 'green', ax)
draw3d(1, 1, 1, 3, 'blue', ax)

ax.set_xlabel('Trục X')
ax.set_ylabel('Trục Y')
ax.set_zlabel('Trục Z')
plt.show()
```

```python
a0, a1, a2 = sp.symbols('a0 a1 a2')

eq1 = sp.Eq(a0 + a1*1 + a2*1**2, 6)
eq2 = sp.Eq(a0 + a1*2 + a2*2**2, 15)
eq3 = sp.Eq(a0 + a1*3 + a2*3**2, 38)

solution = sp.solve((eq1, eq2, eq3), (a0, a1, a2))

print(f"a0 = {solution[a0]}")
print(f"a1 = {solution[a1]}")
print(f"a2 = {solution[a2]}")
```

```python
import sympy as sp

tre_em, nguoi_lon = sp.symbols('c a')

eq1 = sp.Eq(3.0*tre_em + 3.2*nguoi_lon, 118.4)
eq2 = sp.Eq(3.5*tre_em + 3.6*nguoi_lon, 135.2)

solution = sp.solve((eq1, eq2), (tre_em, nguoi_lon))

print(f"So tre em: {round(solution[tre_em])} ")
print(f"So nguoi lon: {round(solution[nguoi_lon])} ")
```

```python
A = np.array([
    [0.25, 0.15, 0.1],
    [0.4, 0.15, 0.2],
    [0.15, 0.2, 0.2]
])
d = np.array([100, 100, 100])

I = np.eye(3)

#   p - Ap = d 
#-> p * (I - A) = d

ve_trai = I - A

p = np.linalg.solve(ve_trai, d)

print(f"(O):   {p[0]:.2f}")
print(f"(E):   {p[1]:.2f}")
print(f"(S):   {p[2]:.2f}")
```

```python

x1, x2, x3, x4 = sp.symbols('x1 x2 x3 x4')

eqC = sp.Eq(3*x1 - x3, 0) # x1 3 carbon = x3 1 carbon
eqH = sp.Eq(8*x1 - 2*x4, 0) # x1 8 hydrogen = x4 2 hydrogen
eqO = sp.Eq(2*x2 - 2*x3 - x4, 0) # x2 2 oxygen = x3 2 oxygen + x4 1 oxygen

solution = sp.solve((eqC, eqH, eqO), (x2, x3, x4))

# do vo so nghiem nen dat he so cua carbon la 1 

he_so_carbon = 1
print(f"x1 = {he_so_carbon}")
print(f"x2 = {solution[x2].subs(x1, he_so_carbon)}")
print(f"x3 = {solution[x3].subs(x1, he_so_carbon)}")
print(f"x4 = {solution[x4].subs(x1, he_so_carbon)}")
```
