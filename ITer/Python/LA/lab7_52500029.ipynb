```python
import numpy as np
import math 
```

```python
def Ex1(A):
    A=abs(A)
    max = sum(A.T[0])
    for col in A.T:
        tmp = sum(col)
        if tmp > max:
            max=tmp
    return max
```

```python
A1 = np.array([[1, -7], [-2, -3]])
print("1-Norm of A1: ", Ex1(A1))

A2 = np.array([[-2, 8], [3, 1]])
print("1-Norm of A2: ", Ex1(A2))

A3 = np.array([[2, -8], [3, 1]])
print("1-Norm of A3: ", Ex1(A3))

A4 = np.array([[2, 3], [1, -1]])
print("1-Norm of A4: ", Ex1(A4))

A5 = np.array([[5, -4, 2], [-1, 2, 3], [-2, 1, 0]])
print("1-Norm of A5: ", Ex1(A5))
```

```python
def Ex2(B):
    B=abs(B)
    max = sum(B[0])
    for row in B:
        tmp = sum(row)
        if tmp > max:
            max=tmp
    return max
```

```python
B1 = np.array([[1, -7], [-2, -3]])
print("\u221E-Norm of B1: ", Ex1(B1))

B2 = np.array([[3, 6], [1, 0]])
print("\u221E-Norm of B2: ", Ex1(B2))

B3 = np.array([[5, -4, 2], [-1, 2, 3], [-2, 1, 0]])
print("\u221E-Norm of B3: ", Ex1(B3))

B4 = np.array([[3, 6, -1], [3, 1, 0], [2, 4, -7]])
print("\u221E-Norm of B4: ", Ex1(B4))

B5 = np.array([[-3, 0, 0], [0, 4, 0], [0, 0, 2]])
print("\u221E-Norm of B5: ", Ex1(B5))
```

```python
def Ex2(C):
    frobenius_norm=0
    C_squared=C**2
    for row in C_squared:
        frobenius_norm += sum(row)
    return math.sqrt(frobenius_norm)
```

```python
C1 = np.array([[5, -4, 2], 
               [-1, 2, 3], 
               [-2, 1, 0]])
print("Frobenius norm của C1: ", Ex2(C1))
C2 = np.array([[1, 7, 3], 
               [4, -2, -2], 
               [-2, -1, 1]])
print("Frobenius norm của C2: ", Ex2(C2))
C3 = np.array([[2, 3], 
               [1, -1]])
print("Frobenius norm của C3: ", Ex2(C3))
```

```python
def angle_between(u, v):
    dot_product = np.dot(u, v)
    
    norm_v1 = np.linalg.norm(u)
    norm_v2 = np.linalg.norm(v)
    
    cos_theta = dot_product / (norm_v1 * norm_v2)
    
    angle_radian = np.arccos(cos_theta)
    angle_degree = np.degrees(angle_radian)
    
    return angle_degree
```

```python
u = np.array([1, 1])
u = np.array([0, 1])
print(f"(a)Góc giữa hai vectơ: {angle_between(u,u):.2f} độ")
u = np.array([1, 0])
u = np.array([0, 1])
print(f"(b)Góc giữa hai vectơ: {angle_between(u,u):.2f} độ")
u = np.array([-2, 3])
u = np.array([1/2, -(1/2)])
print(f"(c)Góc giữa hai vectơ: {angle_between(u,u):.2f} độ")
```

```python
def get_unit_vector(u):
    norm = np.linalg.norm(u)
    if norm == 0:
        return u
    return u / norm
```

```python
u1 = np.array([2,3])
print("(a)\nVector đơn vị của u là",get_unit_vector(u1))
u2 = np.array([1,2,3])
print("(b)\nVector đơn vị của u là",get_unit_vector(u2))
u3 = np.array([1/2,-1/2,1/4])
print("(c)\nVector đơn vị của u là",get_unit_vector(u3))
u4 = np.array([np.sqrt(2),2,- np.sqrt(2), np.sqrt(2)])
print("(d)\nVector đơn vị của u là",get_unit_vector(u4))
```

```python
def find_euclid_distance(v1,v2):
    distance = np.linalg.norm(v1-v2)
    return distance    
v1 = np.array([1,2,3])
v2 = np.array([7,4,3])
v3 = np.array([2,1,9])
print(f"Khoảng cách Euclid giữa v1 và s2 là: {find_euclid_distance(v1, v2):.2f}")
print(f"Khoảng cách Euclid giữa v1 và s3 là: {find_euclid_distance(v1, v3):.2f}")
print(f"Khoảng cách Euclid giữa s2 và s3 là: {find_euclid_distance(v2, v3):.2f}")
```

```python
def Ex8(message,A):
    n= math.ceil(len(message)/3)
    D=np.full((3,n),30)
    index=0
    for j in range(n):
        for i in range(3):
            D[i][j]=ord(message[index])-61
            index+=1
            if index==len(message):
                break
    return A@D
```

```python
def cal_similar(doc_i,doc_j):
    result = np.dot(doc_i,doc_j)/(np.linalg.norm(doc_i)*np.linalg.norm(doc_j))
    return result
doc1 = np.array([0,4,0,0,0,2,1,3])
doc2 = np.array([3,1,4,3,1,2,0,1])
doc3 = np.array([3,0,0,0,3,0,3,0])
doc4 = np.array([0,1,0,3,0,0,2,0])
doc5 = np.array([2,2,2,3,1,4,0,2])
# Tim cosine similarity cua doc 1 va 2
print(cal_similar(doc1,doc2))
```
