```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math

x, m = sp.symbols('x m')
```

```python
# cau 1
def ex1():
    f_a = 3*x**4 - 16*x**3 + 18*x**2 - 9
    f_b = (x + 2)/(2*x**2)
    f_c = -x**2/3 + x**2 + 3*x + 4
    f_d = (5*x**2 + 5)/x
    
    print("Cau 1a:", sp.solve(sp.diff(f_a, x), x))
    print("Cau 1b:", sp.solve(sp.diff(f_b, x), x))
    print("Cau 1c:", sp.solve(sp.diff(f_c, x), x))
    print("Cau 1d:", sp.solve(sp.diff(f_d, x), x))

ex1()
```

```python
# cau 2
def ex2(f, name):
    df = sp.diff(f, x)
    df2 = sp.diff(f, x, 2)
    cps = sp.solve(df, x)
    res = []
    for cp in cps:
        val = df2.subs(x, cp)
        if val > 0: res.append(f"x={cp} (Cuc tieu)")
        elif val < 0: res.append(f"x={cp} (Cuc dai)")
        else: res.append(f"x={cp} (Chua ro)")
    print(f"Cau 2{name}:", ", ".join(res))

ex2(3*x**4 - 16*x**3 + 18*x**2 - 9, 'a')
ex2((x + 2)/(2*x**2), 'b')
ex2(-x**2/3 + x**2 + 3*x + 4, 'c')
ex2((5*x**2 + 5)/x, 'd')
```

```python
# cau 3
def ex3_func(f, a, b, name):
    df = sp.diff(f, x)
    cps = sp.solve(df, x)
    pts = [float(v) for v in cps if v.is_real and a <= v <= b] + [float(a), float(b)]
    vals = [float(f.subs(x, p)) for p in pts]
    print(f"Cau 3{name}: GTLN = {max(vals):.4f}, GTNN = {min(vals):.4f}")

ex3_func(x**3 - 27*x, 0, 5, 'a')
ex3_func(3/2*x**4 - 4*x**3 + 4, 0, 3, 'b')
ex3_func(1/2*x**4 - 4*x**2 + 5, 1, 3, 'c')
ex3_func(5/2*x**4 - 20/3*x**3 + 6, -1, 3, 'd')
```

```python
# cau 4
def draw_ex4(f_expr, a, b, name):
    f_func = sp.lambdify(x, f_expr, 'numpy')
    df = sp.diff(f_expr, x)
    try:
        cps = sp.solve(df, x)
        pts = [float(v) for v in cps if v.is_real and a <= v <= b] + [float(a), float(b)]
    except NotImplementedError:
        pts = np.linspace(float(a), float(b), 1000).tolist()
        
    vals = [float(f_expr.subs(x, p)) for p in pts]
    max_x, max_y = pts[vals.index(max(vals))], max(vals)
    min_x, min_y = pts[vals.index(min(vals))], min(vals)
    
    print(f"Cau 4{name}: GTLN tai x={max_x:.4f} la {max_y:.4f}, GTNN tai x={min_x:.4f} la {min_y:.4f}")
    
    xs = np.linspace(float(a), float(b), 200)
    plt.plot(xs, f_func(xs))
    plt.scatter([max_x, min_x], [max_y, min_y], color=['red', 'blue'])
    plt.title(f"Cau 4{name}")
    plt.show()

draw_ex4(x**2 - 2*x - 5, 0, 2, 'a')
draw_ex4(3*x + x**3 + 5, -4, 4, 'b')
draw_ex4(sp.sin(x) + 3*x**2, -2, 2, 'c')
draw_ex4(sp.exp(x) + 3*x, -1, 1, 'd')
draw_ex4(x**3 - 3*x, -3, 0, 'e')
draw_ex4(x**3 - 3*x, 0, 3, 'f')
draw_ex4(sp.sin(x), 0, float(sp.pi), 'g')
draw_ex4(sp.sin(2*x), 0, 2, 'h')
draw_ex4(sp.cos(x), float(sp.pi/2), float(3*sp.pi/2), 'i')
draw_ex4(sp.tan(x)**2, float(-sp.pi/4), float(sp.pi/4), 'j')
draw_ex4(sp.exp(x)*sp.sin(x), 0, float(sp.pi), 'k')
draw_ex4(x**4 - 3*x**2, -4, 0, 'l')
draw_ex4(x**4 - 3*x**2, 0, 4, 'm')
draw_ex4(x**5 - 5*x**3, -4, 0, 'n')
draw_ex4(x**6 - 5*x**2, -1, 1, 'o')
draw_ex4(x**3 - 9*x, -3, 0, 'p')
draw_ex4(x**3 - 9*x, 0, 3, 'q')
draw_ex4(x**3 + 9*x, -1, 1, 'r')
```

```python
# cau 5
def golden_search(f, a, b, eps=0.3):
    phi = 0.618
    d = b - a
    iterations = []
    while (b - a) >= eps:
        d = phi * d
        x1, x2 = b - d, a + d
        f1, f2 = f(x1), f(x2)
        iterations.append([a, b, x1, x2, f1, f2])
        if f1 <= f2: b = x2
        else: a = x1
    return a, b, iterations

a5, b5, table5 = golden_search(lambda v: v**2, -2, 1, 0.3)
print("Cau 5: Golden Search Table")
print("Iter |     a     |     b     |     x1    |     x2    |   f(x1)   |   f(x2)")
for i, row in enumerate(table5):
    print(f"{i+1:4d} | {row[0]:9.4f} | {row[1]:9.4f} | {row[2]:9.4f} | {row[3]:9.4f} | {row[4]:9.4f} | {row[5]:9.4f}")
print(f"Khoang hoi tu = ({a5:.4f}, {b5:.4f})")

xs = np.linspace(-2, 1, 200)
plt.plot(xs, xs**2, label="f(x)=x^2")
plt.axvline(a5, color='red', linestyle='--', label="Khoang cuoi")
plt.axvline(b5, color='red', linestyle='--')
plt.legend()
plt.title("Cau 5: Golden Search")
plt.show()
```

```python
# cau 6
def fibonacci_search(f, a, b, eps=0.3):
    f_prev, f_curr = 2, 3
    iterations = []
    while (b - a) >= eps:
        d = b - a
        ratio = f_prev / f_curr
        x1, x2 = b - d * ratio, a + d * ratio
        f1, f2 = f(x1), f(x2)
        iterations.append([a, b, x1, x2, f1, f2])
        if f1 <= f2: b = x2
        else: a = x1
        f_prev, f_curr = f_curr, f_prev + f_curr
    return a, b, iterations

a6, b6, table6 = fibonacci_search(lambda v: v**2, -2, 1, 0.3)
print("Cau 6: Fibonacci Search Table")
print("Iter |     a     |     b     |     x1    |     x2    |   f(x1)   |   f(x2)")
for i, row in enumerate(table6):
    print(f"{i+1:4d} | {row[0]:9.4f} | {row[1]:9.4f} | {row[2]:9.4f} | {row[3]:9.4f} | {row[4]:9.4f} | {row[5]:9.4f}")
print(f"Khoang hoi tu = ({a6:.4f}, {b6:.4f})")

xs = np.linspace(-2, 1, 200)
plt.plot(xs, xs**2, label="f(x)=x^2")
plt.axvline(a6, color='red', linestyle='--', label="Khoang cuoi")
plt.axvline(b6, color='red', linestyle='--')
plt.legend()
plt.title("Cau 6: Fibonacci Search")
plt.show()
```

```python
# cau 7
def ex7():
    y = x**3 - 3*m*x**2 + 3*(m**2 - 1)*x - (m**2 - 1)
    y1 = sp.diff(y, x)
    y2 = sp.diff(y, x, 2)
    
    sols = sp.solve(y1.subs(x, 1), m)
    res = []
    for sol in sols:
        if y2.subs({x: 1, m: sol}) < 0:
            res.append(str(sol))
    print("Cau 7: Cac gia tri m thoa man =", ", ".join(res))

ex7()
```

```python
# cau 8
def golden_search_auto(f, a, b, eps=0.01):
    mid = (a + b) / 2
    h = 1e-5
    curv = (f(mid+h) - 2*f(mid) + f(mid-h)) / (h*h)
    mode = "min" if curv > 1e-6 else "max" if curv < -1e-6 else "auto"
    
    phi = 0.618
    while (b - a) >= eps:
        d = phi * (b - a)
        x1, x2 = b - d, a + d
        f1, f2 = f(x1), f(x2)
        
        if mode == "min":
            if f1 <= f2: b = x2
            else: a = x1
        elif mode == "max":
            if f1 >= f2: b = x2
            else: a = x1
        else:
            if f1 > f2: a = x1
            else: b = x2
    return (a + b) / 2, mode

def draw_ex8(f, a, b, eps, name):
    opt, mode = golden_search_auto(f, a, b, eps)
    print(f"Cau 8{name}: {mode} tai {opt:.4f}, f = {f(opt):.4f}")
    xs = np.linspace(a, b, 200)
    ys = [f(v) for v in xs]
    plt.plot(xs, ys)
    plt.scatter([opt], [f(opt)], color='red')
    plt.title(f"Cau 8{name}")
    plt.show()

draw_ex8(lambda v: -2*v**2 + v + 4, -5, 5, 1/9, 'a')
draw_ex8(lambda v: -4*v**2 + 2*v + 2, -6, 6, 1/10, 'b')
draw_ex8(lambda v: v**3 + 6*v**2 + 5*v - 12, -5, -2, 1/10, 'c')
draw_ex8(lambda v: 2*v - v**2, 0, 3, 1/100, 'd')
draw_ex8(lambda v: v**2 - v - 10, -10, 10, 1/5, 'e')
draw_ex8(lambda v: -(v+6)**2 + 4, -10, 10, 1/8, 'f')
draw_ex8(lambda v: -2*v**2 + 3*v + 6, -3, 5, 1/8, 'g')
```
