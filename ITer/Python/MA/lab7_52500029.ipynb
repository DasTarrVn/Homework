```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n, x = sp.symbols('n x')

# cau 1
def ex1(num):
    # cau 1a
    print("Cau 1a:", [4*i + 1 for i in range(1, num+1)])
    # cau 1b
    print("Cau 1b:", [3**i for i in range(1, num+1)])
    # cau 1c
    print("Cau 1c:", [i**3 for i in range(1, num+1)])
    # cau 1d
    f = [1, 1]
    for _ in range(2, num): f.append(f[-1] + f[-2])
    print("Cau 1d:", f[:num])
ex1(10)
```

```python
# cau 3
print("Cau 3a:", sp.series(sp.cos(x), x, sp.pi/3, 6).removeO())
print("Cau 3b:", sp.series(sp.ln(x), x, 2, 10).removeO())
print("Cau 3c:", sp.series(sp.exp(x), x, 3, 12).removeO())
```

```python
# cau 4
print("Cau 4a:", sp.series(sp.cos(x), x, 0, 6).removeO())
print("Cau 4b:", sp.series(sp.exp(x), x, 0, 12).removeO())
print("Cau 4c:", sp.series(1/(1-x), x, 0, 12).removeO())
print("Cau 4d:", sp.series(sp.atan(x), x, 0, 12).removeO())
```

```python
# cau 5
print("Cau 5a:", sp.limit((4*n**2 + 1)/(3*n**2 + 2), n, sp.oo))
print("Cau 5b:", sp.limit(sp.sqrt(n**2 + 1) - n, n, sp.oo))
print("Cau 5c:", sp.limit(sp.sqrt(2*n + sp.sqrt(n)) - sp.sqrt(2*n + 1), n, sp.oo))
print("Cau 5d:", sp.limit((3*5**n - 2**n)/(4**n + 2*5**n), n, sp.oo))
print("Cau 5e:", sp.limit((n * sp.sin(sp.sqrt(n)))/(n**2 + n - 1), n, sp.oo))
```

```python
# cau 6
def check_conv(expr, name):
    L = sp.limit(expr, n, sp.oo)
    if L.is_finite: print(f"Cau 6{name}: Hoi tu ve", L)
    else: print(f"Cau 6{name}: Phan ky")
check_conv(1 - 0.2**n, 'a')
check_conv(n**3 / (n**3 + 1), 'b')
check_conv((3 + 5*n**2) / (n + n**2), 'c')
check_conv(n**3 / (n + 1), 'd')
check_conv(sp.exp(1/n), 'e')
check_conv(sp.sqrt((n + 1) / (9*n + 1)), 'f')
try:
    lim_g = sp.limit((-1)**(n+1) * n / (n + sp.sqrt(n)), n, sp.oo)
    print(f"Cau 6g: Hoi tu ve {lim_g}" if lim_g.is_finite else "Cau 6g: Phan ky")
except:
    print("Cau 6g: Phan ky (giao dong)")
check_conv(sp.tan(2*n*sp.pi / (1 + 8*n)), 'h')
check_conv(sp.factorial(2*n - 1) / sp.factorial(2*n + 1), 'i')
check_conv(sp.ln(2*n**2 + 1) - sp.ln(n**2 + 1), 'j')
```

```python
# cau 7
def draw_ex7(terms, name):
    print(f"Cau 7{name}: 5 so hang dau =", terms)
    plt.plot(range(1, 6), terms, marker='o')
    plt.title(f"Cau 7{name}")
    plt.show()

draw_ex7([1 - 0.2**i for i in range(1, 6)], 'a')
draw_ex7([2*i / (i**2 + 1) for i in range(1, 6)], 'b')
draw_ex7([((-1)**(i-1)) / (5**i) for i in range(1, 6)], 'c')
draw_ex7([1 / math.factorial(i+1) for i in range(1, 6)], 'd')

# cau 7e
te = [1]
for _ in range(4): te.append(5*te[-1] - 3)
draw_ex7(te, 'e')

# cau 7f
tf = [2]
for _ in range(4): tf.append(tf[-1] / (tf[-1] + 1))
draw_ex7(tf, 'f')
```

```python
# cau 8
def draw_ex8(func, name, n_val=50):
    t = [func(i) for i in range(1, n_val + 1)]
    plt.plot(range(1, n_val + 1), t, marker='.')
    plt.title(f"Cau 8{name}")
    plt.show()

draw_ex8(lambda i: 1 - (-2/math.e)**i, 'a')
draw_ex8(lambda i: math.sqrt(i) * math.sin(math.pi / math.sqrt(i)), 'b')
draw_ex8(lambda i: math.sqrt((3 + 2*i**2) / (8*i**2 + i)), 'c')
draw_ex8(lambda i: i**2 * math.cos(i) / (1 + i**2), 'd')
draw_ex8(lambda i: math.prod(range(1, 2*i, 2)) / math.factorial(i), 'e', 15)
draw_ex8(lambda i: math.prod(range(1, 2*i, 2)) / ((2*i)**i), 'f', 15)
```

```python
# cau 9
s9a = sp.Sum(4**n, (n, 1, sp.oo))
print("Cau 9a:", "Hoi tu" if s9a.is_convergent() else "Phan ky")
s9b = sp.Sum(5 / (2**n), (n, 1, sp.oo))
print("Cau 9b:", "Hoi tu ve " + str(s9b.doit()) if s9b.is_convergent() else "Phan ky")
```

```python
# cau 10
def fibo10(i):
    # cau 10a
    if i <= 2: fa = 1
    else:
        arr = [1, 1]
        for _ in range(2, i): arr.append(arr[-1] + arr[-2])
        fa = arr[-1]
    
    # cau 10b
    phi = 1.618034
    fb = round((phi**i - (1 - phi)**i) / math.sqrt(5))
    
    # cau 10c
    fc = 1
    for _ in range(2, i+1): fc = round(fc * phi)
    print(f"Cau 10 (tim so thu i={i}): cach a={fa}, cach b={fb}, cach c={fc}")

fibo10(10)
```
