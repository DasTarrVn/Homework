```python
import sympy as sp 
import math 

x = sp.symbols('x')

f5a = abs (x )** (1/2)
sp.plot(f5a , (x,-10,10 ), line_color='red')

f5d = math.e**x 
sp.plot (f5d , (x,-10,10 ),line_color='blue')

f5e = sp.log(x)
sp.plot(f5e , (x,-10,10))
```

```python
import sympy as sp 
import matplotlib.pyplot 
import numpy as np 
x = np.arange(-10,10,0.1 )
f1=lambda x:-x+5
f2=lambda x:x/2 +2

y1=f1(x)
y2=f2(x)

matplotlib.pyplot.plot(x,y1)
matplotlib.pyplot.plot(x,y2)

x = sp.symbols('x')
f1_sym=f1(x)
f2_sym=f2(x)
x_root=sp.solve(f1_sym-f2_sym)
y_root=f1_sym.subs(x,x_root[0])
matplotlib.pyplot.plot(x_root,y_root,'ro')
matplotlib.pyplot.title('Find intersection point of f1(x) and f2(x)')
matplotlib.pyplot.grid(linestyle='--')
matplotlib.pyplot.show()
```
