```python
x=sp.symbols('x')
f=abs(x**2 - x - 7 )
result= sp.limit(f,x,3)
print(result)
#cau 2
x_num = np.linspace(0, 6, 100) 
y_num = np.abs(x_num**2 - x_num - 7)
plt.plot(x_num, y_num, label='f(x)', color='blue')
plt.plot(3, result, 'ro', label='Limit point (3, 1)') 

plt.title("Graph of (a) and limit point")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend() 
plt.grid(True)

plt.show()
```

```python
import sympy as sp

x = sp.symbols('x')

print("--- Part (a) ---")
f_a = x**2 - 7
c_a = 1

value_a = f_a.subs(x, c_a)
limit_a = sp.limit(f_a, x, c_a)

print(f"f({c_a}) = {value_a}")
print(f"Limit as x->{c_a} = {limit_a}")

if value_a == limit_a:
    print("Conclusion: The function is continuous at c = 1")
else:
    print("Conclusion: The function is discontinuous at c = 1")

print("\n--- Part (b) ---")
f_b = sp.sqrt(2*x - 3)
c_b = 2

value_b = f_b.subs(x, c_b)
limit_b = sp.limit(f_b, x, c_b)

print(f"f({c_b}) = {value_b}")
print(f"Limit as x->{c_b} = {limit_b}")

if value_b == limit_b:
    print("Conclusion: The function is continuous at c = 2")
else:
    print("Conclusion: The function is discontinuous at c = 2")
```

```python
x = sp.symbols('x')

print("--- Function 1 ---")
f1 = (x**2 - x - 2) / (x - 2)

denominator1 = x - 2
points1 = sp.solve(denominator1, x)

print(f"Denominator is zero at: {points1}")
print(f"Conclusion: Function is discontinuous at x = {points1[0]}")

limit1 = sp.limit(f1, x, points1[0])
print(f"Limit at this point: {limit1}")

print("\n--- Function 2 ---")
f2 = (x**2 - 2*x - 3) / (2*x - 6)

denominator2 = 2*x - 6
points2 = sp.solve(denominator2, x)

print(f"Denominator is zero at: {points2}")
print(f"Conclusion: Function is discontinuous at x = {points2[0]}")

limit2 = sp.limit(f2, x, points2[0])
print(f"Limit at this point: {limit2}")
```

```python
x = sp.symbols('x')
f = 1 - sp.sqrt(1 - x**2)

print("--- Check Left Endpoint (x = -1) ---")
value_left = f.subs(x, -1)
limit_left = sp.limit(f, x, -1, dir='+')

print(f"f(-1) = {value_left}")
print(f"Limit x->-1+ = {limit_left}")

is_continuous_left = (value_left == limit_left)

print("\n--- Check Right Endpoint (x = 1) ---")
value_right = f.subs(x, 1)
limit_right = sp.limit(f, x, 1, dir='-')

print(f"f(1) = {value_right}")
print(f"Limit x->1- = {limit_right}")

is_continuous_right = (value_right == limit_right)

print("\n--- Conclusion ---")
if is_continuous_left and is_continuous_right:
    print("The function is continuous on the interval [-1, 1]")
else:
    print("The function is NOT continuous on the interval [-1, 1]")
```
