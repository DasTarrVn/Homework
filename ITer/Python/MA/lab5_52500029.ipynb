```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python

x, t, z = sp.symbols('x t z')
funcs = {
    "a": (4 - x**2, x),
    "b": ((x - 1)**2 + 1, x),
    "c": (1 / t**2, t),
    "d": ((1 - z) / (2 * z), z),
    "e": (sp.sqrt(3 * z), z),
    "f": (sp.sqrt(2 * z + 1), z)
}
for k, (f, v) in funcs.items():
    print(f"({k}) {sp.simplify(sp.diff(f, v))}")
```

```python
x, t, z = sp.symbols('x t z')
funcs = {
    "a": (x**2 + 1, x, 2, 5),
    "b": (x - 2*x**2, x, 1, -1),
    "c": (x/(x-2), x, 3, 3),
    "d": (8/x**2, x, 2, 2),
    "e": (sp.sqrt(x), x, 4, 2),
    "f": (t**3 + 3*t, t, 1, 4),
    "g": (8/sp.sqrt(x-2), x, 6, 4),
    "h": (1 + sp.sqrt(4-z), z, 3, 2)
}

for k, (f, var, x0, y0) in funcs.items():
    m = sp.diff(f, var).subs(var, x0)
    print(f"({k}) y = {sp.simplify(m * (var - x0) + y0)}")

x_vals = np.linspace(-1, 5, 400)
f_num = sp.lambdify(x, funcs["a"][0], 'numpy')
tan_num = sp.lambdify(x, 4*(x - 2) + 5, 'numpy')

plt.plot(x_vals, f_num(x_vals))
plt.plot(x_vals, tan_num(x_vals))
plt.scatter([2], [5])
plt.show()
```

```python
x = sp.symbols('x')
funcs = {
    "a": (5*x - 3*x**2, 1),
    "b": (1 / (x - 1), 3),
    "c": (x**3 - 2*x + 7, -2),
    "d": ((x - 1) / (x + 1), 0)
}

for k, (f, x0) in funcs.items():
    m = sp.diff(f, x).subs(x, x0)
    print(f"({k}) m = {m}, y = {sp.simplify(m * (x - x0) + f.subs(x, x0))}")
```

```python
x, dx = sp.symbols('x dx')
f = -2*x**2 / 3 + x
print(sp.limit((f.subs(x, dx) - f.subs(x, 0)) / dx, dx, 0))
```

```python
x, t, z, dx = sp.symbols('x t z dx')
data = {
    "a": (4 - x**2, x, [-3, 0, 1]),
    "b": ((x - 1)**2 + 1, x, [-1, 0, 2]),
    "c": (1 / t**2, t, [-1, 2, sp.sqrt(3)]),
    "d": ((1 - z) / (2 * z), z, [-1, 1, sp.sqrt(2)])
}

for k, (f, v, pts) in data.items():
    deriv = sp.limit((f.subs(v, v + dx) - f) / dx, dx, 0)
    res = ", ".join([str(sp.simplify(deriv.subs(v, p))) for p in pts])
    print(f"({k}) {res}")
```
