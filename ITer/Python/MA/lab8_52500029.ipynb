```python
import sympy as sp
from sympy.plotting import plot3d
```

```python
x, y, z, t, h = sp.symbols('x y z t h')

# cau 1
f1a = x**2 + x*y**3
print("Cau 1a:", [f1a.subs({x: p[0], y: p[1]}) for p in [(0,0), (-1,1), (2,3), (-3,-2)]])

f1b = (x - y)/(y**2 + z**2)
print("Cau 1b:", [f1b.subs({x: p[0], y: p[1], z: p[2]}) for p in [(3,-1,2), (1, 1/2, 1/4), (0, -1/3, 0), (2, 2, 100)]])
```

```python
# cau 2
f2 = [
    sp.cos(x)*sp.cos(y)*sp.exp(-sp.sqrt(x**2+y**2)/4),
    -x*y**2 / (x**2+y**2),
    x*y*(x**2-y**2) / (x**2+y**2),
    y**2 - y**4 - x**2
]
for i, f in enumerate(f2):
     plot3d(f, (x, -5, 5), (y, -5, 5), title=f"Cau 2{chr(97+i)}")
```

```python
# cau 3
f3 = [
    2*x**2 - 3*y - 4, (x**2 - 1)*(y + 2), x**2 - x*y + y**2,
    5*x*y - 7*x**2 - y**2 + 3*x - 6*y + 2, (x*y - 1)**2, (2*x - 3*y)**3,
    sp.sqrt(x**2 + y**2), (x**3 + y/2)**(2/3), 1/(x+y), x/(x**2+y**2),
    (x+y)/(x*y-1), sp.atan(y/x), sp.exp(x+y+1), sp.exp(-x)*sp.sin(x+y), sp.ln(x+y)
]

c = input("Nhap cau 3 (a-o): ").strip().lower()
i = ord(c) - 97

if 0 <= i < len(f3):
    f = f3[i]
    print(f"Cau 3{c}:\ndf/dx = {sp.diff(f, x)}\ndf/dy = {sp.diff(f, y)}")
    try: 
        plot3d(f, sp.diff(f, x), sp.diff(f, y), (x, -5, 5), (y, -5, 5))
    except: 
        print("Loi ve do thi")
```

```python
# cau 4
f4 = [
    x+y+x*y, sp.sin(x*y), x**2*y + sp.cos(y) + y*sp.sin(x),
    x*sp.exp(y) + y + 1, sp.ln(x+y), sp.atan(y/x),
    x**2*sp.tan(x*y), y*sp.exp(x**2 - y), x*sp.sin(x**2*y), (x-y)/(x**2+y)
]

c = input("Nhap cau 4(a-j): ").strip().lower()
i = ord(c) - 97

if 0 <= i < len(f4):
    f = f4[i]
    print(f"Cau 4{c}:\nf_xx = {sp.diff(f, x, 2)}\nf_yy = {sp.diff(f, y, 2)}\nf_xy = {sp.diff(f, x, y)}\nf_yx = {sp.diff(f, y, x)}")
    try: 
        plot3d(f, sp.diff(f, x, 2), sp.diff(f, y, 2), (x, -5, 5), (y, -5, 5))
    except: 
        print("Loi ve do thi")
```

```python
# cau 5
f5 = [
    x*sp.sin(y) + y*sp.sin(x) + x*y, sp.ln(2*x + 3*y),
    x*y**2 + x**2*y**3 + x**3*y**4, sp.exp(x) + x*sp.ln(y) + y*sp.ln(x)
]
for i, f in enumerate(f5):
    print(f"Cau 5{chr(97+i)}:", "Bang nhau" if sp.simplify(sp.diff(f, x, y) - sp.diff(f, y, x)) == 0 else "Khong bang nhau")
```

```python
# cau 6
f6 = [
    y**2 * x**4 * sp.exp(x) + 2,
    y**4 + y*(sp.sin(x) - x**4),
    x**5 + 5*x*y + sp.sin(x) + 7*sp.exp(x),
    x*sp.exp(y**4 / 2)
]
for i, f in enumerate(f6):
    print(f"Cau 6{chr(97+i)}: {sp.diff(f, x, 2, y, 3)}")
```

```python
# cau 7
def solve_ex7(w, sub_dict, t_val):
    return sp.simplify(sp.diff(w.subs(sub_dict), t).subs({t: t_val}))

print("Cau 7a:", solve_ex7(x**2 + y**2, {x: sp.cos(t), y: sp.sin(t)}, sp.pi))
print("Cau 7b:", solve_ex7(x**2 + y**2, {x: sp.cos(t)+sp.sin(t), y: sp.cos(t)-sp.sin(t)}, 0))
print("Cau 7c:", solve_ex7(x/z + y/z, {x: sp.cos(t)**2, y: sp.sin(t)**2, z: 1/t}, 3))
print("Cau 7d:", solve_ex7(2*y*sp.exp(x) - sp.ln(z), {x: sp.ln(t**2+1), y: sp.atan(t), z: sp.exp(t)}, 1))
print("Cau 7e:", solve_ex7(z - sp.sin(x*y), {x: t, y: sp.ln(t), z: sp.exp(t-1)}, 1))
```

```python
# cau 8
def limit_diff(f, pt):
    fx = sp.limit((f.subs({x: pt[0]+h, y: pt[1]}) - f.subs({x: pt[0], y: pt[1]}))/h, h, 0)
    fy = sp.limit((f.subs({x: pt[0], y: pt[1]+h}) - f.subs({x: pt[0], y: pt[1]}))/h, h, 0)
    return fx, fy

print("Cau 8a:", limit_diff(1 - x + y - 3*x**2*y, (1, 2)))
print("Cau 8b:", limit_diff(4 + 2*x - 3*y - 3*x*y**2, (-2, 1)))
```
