```python
import math
#Given x 
x=float(input("Enter x"))
def func_a(x):
    if x < 0:
        return None
    return math.sqrt(x)

def func_b(x):
    if x>=0:
        return x**(1/3)
    else:
        return -(-x)**(1/3)
    
def func_c(x):
    return abs(x)**(2/3)

def func_d(x):
    return x**3/3-x**2/2-2*x+1/3

def func_e(x):
    if (7*x+4)==0:
        return None
    return (2*x**2-3)/(7*x+4)

def func_f(x):
    return (5*x**2+8*x-3)/(3*x**2+2)

def func_g(x):
    return math.sin(x)

def func_h(x):
    return math.cos(x)

def func_i(x):
    return 3**x

def func_j(x):
    return 10**(-x)

def func_k(x):
    return math.exp(x)

def func_l(x):
    if x<=0: return None
    return math.log2(x)

def func_m(x):
    if x<=0: return None
    return math.log10(x)

def func_n(x):
    if x<=0: return None
    return math.log(x)
#Test (Round to 5 decimal)
print("Câu a: ",round(func_a(x),5))
print("Câu b: ",round(func_b(x),5))
print("Câu c: ",round(func_c(x),5))
print("Câu d: ",round(func_d(x),5))
print("Câu e: ",round(func_e(x),5))
print("Câu f: ",round(func_f(x),5))
print("Câu g: ",round(func_g(x),5))
print("Câu h: ",round(func_h(x),5))
print("Câu i: ",round(func_i(x),5))
print("Câu j: ",round(func_j(x),5))
print("Câu k: ",round(func_k(x),5))
print("Câu l: ",round(func_l(x),5))
print("Câu m: ",round(func_m(x),5))
print("Câu n: ",round(func_n(x),5))
```

```python
import numpy as np
import math 
def fx_a(x):
    return 2+ x**2/(x*x+4)
def fx_b(x):
    return math.sqrt(5*x+10)
def fx_c(x):
    return 2/(x**2-16)
def fx_d(x):
    return x**4 +3*x**2 -1
def fx_e(x):
    if x>=0:
        return x
    else:
        return -x

def solve_min_max(fx_name,start,end):
    fx = []
    for x in np.arange(start,end+0.1,0.1):
        y = fx_name(x)
        fx.append(y)
    print(f"--- Result of function {fx_name.__name__} ---")
    print("Min value: ",round(min(fx),5))
    print("Max value: ",round(max(fx),5))
#Test
solve_min_max(fx_a,-2,2)
solve_min_max(fx_b,0,5)
solve_min_max(fx_c,5,10)
solve_min_max(fx_d,-3,3)
solve_min_max(fx_e,-3,3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(15, 10))
x = np.linspace(-5, 5, 1000)
x_with_nan = x.copy()
x_with_nan[np.abs(x_with_nan) < 0.01] = np.nan 
plt.subplot(2, 3, 1)
y_i = -x**3
plt.plot(x, y_i, 'tab:blue')
plt.title("(i) $f(x) = -x^3$")
plt.grid(True)
plt.subplot(2, 3, 2)
y_j = -1 / (x_with_nan**2)
plt.plot(x_with_nan, y_j, 'tab:orange')
plt.title("(j) $f(x) = -1/x^2$")
plt.grid(True)
plt.ylim(-10, 1) 
plt.subplot(2, 3, 3)
y_k = -1 / x_with_nan
plt.plot(x_with_nan, y_k, 'tab:green')
plt.title("(k) $f(x) = -1/x$")
plt.grid(True)
plt.ylim(-10, 10)
plt.subplot(2, 3, 4)
y_l = 1 / np.abs(x_with_nan)
plt.plot(x_with_nan, y_l, 'tab:red')
plt.title("(l) $f(x) = 1/|x|$")
plt.grid(True)
plt.ylim(-1, 10)
plt.subplot(2, 3, 5)
y_m = np.sqrt(np.abs(x))
plt.plot(x, y_m, 'tab:purple')
plt.title("(m) $f(x) = \sqrt{|x|}$")
plt.grid(True)
plt.subplot(2, 3, 6)
y_n = np.sqrt(np.abs(-x))
plt.plot(x, y_n, 'tab:brown')
plt.title("(n) $f(x) = \sqrt{|-x|}$")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 1000)
y1 = 1 - (np.abs(x) - 1)**2
y2 = -3 * np.sqrt(1 - np.sqrt(np.abs(x) / 2))
plt.figure(figsize=(8, 8)) 
plt.plot(x, y1, color='magenta', label=r'$f_1(x) = 1 - (|x| - 1)^2$')
plt.plot(x, y2, color='red', label=r'$f_2(x) = -3\sqrt{1 - \sqrt{\frac{|x|}{2}}}$')
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()     
plt.grid(True)   
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
x_a = np.linspace(-10, 10, 200)
k_values_a = [2, 4, 6, 8, 10, 12]
for k in k_values_a:
    y = x_a**2 + k
    plt.plot(x_a, y, label=f'k={k}')
plt.title("(a) $f(x) = x^2 + k$ (Dịch lên)")
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 2)
x_b = np.linspace(-25, 5, 200) 
k_values_b = [2, 4, 6, 8, 10, 12]
for k in k_values_b:
    y = (x_b + k)**2
    plt.plot(x_b, y, label=f'k={k}')
plt.title("(b) $f(x) = (x + k)^2$ (Dịch trái)")
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 3)
x_c = np.linspace(0, 50, 400) 
k_values_c = [1/3, 1, 3, 6]
for k in k_values_c:
    y = k * np.sqrt(x_c)
    plt.plot(x_c, y, label=f'k={k:.2f}')
plt.title("(c) $f(x) = k\sqrt{x}$ (Co giãn dọc)")
plt.legend()
plt.grid(True)
plt.subplot(2, 2, 4)
x_d = np.linspace(-3, 3, 200)
y_org = x_d**3
y_new = (x_d + 1)**3 - 1
plt.plot(x_d, y_org, 'k--', label='Gốc: $x^3$')
plt.plot(x_d, y_new, 'r', label='Biến đổi: $(x+1)^3 - 1$')
plt.title("(d) Dịch trái 1, Xuống 1")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f_a(x):
    return x**3 - x/2

def f_b(x):
    return x**2 + x/2

x = np.linspace(-1.5, 1.5, 400)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x, f_a(x), label=r'$f(x) = x^3 - \frac{x}{2}$')
plt.axhline(y=0, color='red', linestyle='--', label='y=0 (Cắt 3 điểm)')
plt.title("Câu (a): KHÔNG phải hàm đơn ánh")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, f_b(x), label=r'$f(x) = x^2 + \frac{x}{2}$')
plt.axhline(y=0.5, color='red', linestyle='--', label='y=0.5 (Cắt 2 điểm)')
plt.title("Câu (b): KHÔNG phải hàm đơn ánh")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("KẾT LUẬN:")
print("- Hàm (a) cắt đường y=0 tại 3 điểm -> Không phải one-to-one.")
print("- Hàm (b) là Parabol, cắt đường y=0.5 tại 2 điểm -> Không phải one-to-one.")
```
